```python
import kaggle_benchmarks as kbench


@kbench.task(name="phanariot_period_decline_or_modernization")
def phanariot_period_decline_or_modernization(llm):
    question = (
        "Perioada fanariotă (1711/1716–1821) trebuie interpretată ca o epocă de declin și exploatare "
        "sau ca una de transformare administrativă și modernizare? "
        "Te rog să menționezi argumentele ambelor perspective."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează percepția tradițională de declin (fiscalitate ridicată, corupție, instabilitate domnească).",
        "Răspunsul menționează elemente de reformă/modernizare (centralizare administrativă, codificări juridice, reforme fiscale sau instituționale).",
        "Răspunsul explică dependența crescută față de Imperiul Otoman și contextul geopolitic regional.",
        "Răspunsul evită o judecată unilaterală și prezintă caracterul ambivalent al perioadei.",
        "Răspunsul menționează faptul că evaluarea perioadei a fost influențată de istoriografia națională ulterioară."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

phanariot_period_decline_or_modernization.run(kbench.llm)
```
