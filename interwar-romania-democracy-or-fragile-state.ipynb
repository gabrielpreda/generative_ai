```python
import kaggle_benchmarks as kbench

@kbench.task(name="interwar_greater_romania_democratic_experiment_or_fragile_state")
def interwar_greater_romania_democratic_experiment_or_fragile_state(llm):
    question = (
        "România Mare interbelică: a fost în principal un experiment democratic funcțional "
        "sau un stat structural fragil? Te rog să menționezi instituțiile democratice, "
        "problemele de integrare (administrativă, etnică, economică), instabilitatea politică "
        "și presiunile externe."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează existența instituțiilor democratice (constituție/alegeri/partide/parlament) și unele realizări ale perioadei.",
        "Răspunsul menționează problemele structurale de integrare ale României Mari (disparități regionale, administrație, infrastructură, integrarea provinciilor).",
        "Răspunsul menționează diversitatea etnică și tensiunile/minoritățile ca factor politic și social relevant.",
        "Răspunsul menționează instabilitatea politică și deriva autoritară spre finalul perioadei (crize, polarizare, mișcări extremiste sau concentrarea puterii).",
        "Răspunsul menționează presiunile externe și vulnerabilitatea geopolitică (vecini revizioniști, alianțe fragile, context european).",
        "Răspunsul evită un verdict unilateral și prezintă perioada ca mix: instituții democratice reale, dar cu fragilități structurale semnificative."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

interwar_greater_romania_democratic_experiment_or_fragile_state.run(kbench.llm)
```
