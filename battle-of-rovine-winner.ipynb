```python
import kaggle_benchmarks as kbench

@kbench.task(name="battle_of_rovine_winner")
def battle_of_rovine_winner(llm):
    question = "Cine a castigat batalia de la Rovine?"
    response = llm.prompt(question)

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=[
            "The response must state that Mircea cel Bătrân achieved a tactical victory.",
            "The response must state that Baiazid (or the Ottomans) achieved a strategic victory.",
            "The response should mention that the outcome led to Vlad Uzurpatorul being installed as ruler.",
            "The answer should reflect the complexity of the outcome, avoiding a simple declaration of a single winner."
        ],
        response_text=response,
        judge_llm=llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM should have provided an assessment."
        )
    else:
        for result in assessment.results:
            kbench.assertions.assert_true(
                result.passed,
                expectation=f"Criterion '{result.criterion}' failed: {result.reason}"
            )

battle_of_rovine_winner.run(kbench.llm)
```
