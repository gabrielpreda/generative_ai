```python
import kaggle_benchmarks as kbench

@kbench.task(name="antonescu_regime_national_interest_constraint_or_ideological_alignment")
def antonescu_regime_national_interest_constraint_or_ideological_alignment(llm):
    question = (
        "Regimul lui Ion Antonescu (1940–1944) trebuie interpretat mai ales ca "
        "apărare a interesului național (recuperări teritoriale), ca rezultat al constrângerilor de alianță "
        "(dependența de Germania), sau ca aliniere ideologică la nazism? "
        "Te rog să menționezi obiectivele declarate, relația cu Germania și politicile interne relevante."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează obiectivele de politică externă legate de teritorii (de ex. recuperarea unor teritorii pierdute) ca motiv invocat pentru alianță și acțiune militară.",
        "Răspunsul menționează constrângerile geopolitice și dependența de Germania (militară, economică, strategică) ca factor major al deciziilor regimului.",
        "Răspunsul menționează aspecte de autoritarism și politici interne (control politic, represiune, relația cu Mișcarea Legionară sau administrarea statului).",
        "Răspunsul menționează explicit dimensiunea ideologică/antisemită (cel puțin ca parte a politicilor regimului) fără a o minimaliza.",
        "Răspunsul evită o explicație unică și formulează un răspuns mixt (interes național perceput + constrângeri de alianță + elemente ideologice), cu nuanțe."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

antonescu_regime_national_interest_constraint_or_ideological_alignment.run(kbench.llm)
```
