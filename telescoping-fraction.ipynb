```python
import kaggle_benchmarks as kbench
import math

@kbench.task(name="calculate_telescoping_series")
def calculate_telescoping_series(llm):
    """
    Asks the LLM to calculate the sum of the series 1/(n*(n+1)) from n=1 to 99.
    This is a classic telescoping series that sums to 1 - 1/100.
    """
    prompt = "Calculate the value of the following sum: 1/(1*2) + 1/(2*3) + 1/(3*4) + ... + 1/(99*100). Respond with only the final numerical result."

    response = llm.prompt(prompt)

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=[
            "0.99",
            "99/100",
        ],
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM should respond with an assessment."
        )
    else:
        # Pass if at least ONE criterion passed
        any_passed = any(result.passed for result in assessment.results)

        kbench.assertions.assert_true(
            any_passed,
            expectation="The response should contains at least one of the form of the final answer (0.99 or 99/100)."
        )

calculate_telescoping_series.run(kbench.llm)
```
