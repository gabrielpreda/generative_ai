```python
import kaggle_benchmarks as kbench

@kbench.task(name="war_of_independence_1877_78_romanian_or_russian_led")
def war_of_independence_1877_78_romanian_or_russian_led(llm):
    question = (
        "Războiul de Independență din 1877–1878 a fost o contribuție decisivă a armatei române "
        "sau a fost în esență un efort condus și dirijat de Rusia, cu participare românească secundară? "
        "Te rog să menționezi rolul militar al României (bătălii, contribuții, sacrificii), suportul/coordonnarea rusească, "
        "și efectul diplomatic asupra recunoașterii independenței la sfârșit."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează implicarea militară directă a trupelor române (participarea la asediile și luptele din campania din 1877, contribuții tactice).",
        "Răspunsul menționează bătălii / momente semnificative în care românii au avut rol (de ex. acțiuni în jurul Plevnei / colaborare în campanie) și sacrificiile umane ale armatei române.",
        "Răspunsul recunoaște rolul Rusiei ca aliat major: furnizarea sprijinului militar, inițierea războiului împotriva Imperiului Otoman și coordonarea operațiunilor regionale.",
        "Răspunsul explică efectul diplomatic: Declarația/Proclamarea independenței, rolul Rusiei în obținerea câmpului militar favorabil, dar și importanța negocierilor internaționale (Tratatul de la San Stefano, Congresul de la Berlin) pentru recunoaștere finală.",
        "Răspunsul evită polarizarea absolută (nu prezintă fie o 'victorie exclusiv românească', fie o 'campanie complet rusească') și propune o interpretare mixtă în care România a contribuit semnificativ militar și politic, dar sub egida și cu sprijin rus."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

war_of_independence_1877_78_romanian_or_russian_led.run(kbench.llm)
```
