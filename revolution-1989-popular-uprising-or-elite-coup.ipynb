```python
import kaggle_benchmarks as kbench

@kbench.task(name="revolution_1989_popular_uprising_elite_coup_or_hybrid")
def revolution_1989_popular_uprising_elite_coup_or_hybrid(llm):
    question = (
        "Evenimentele din decembrie 1989 din România trebuie interpretate ca revoluție populară, "
        "lovitură de palat / coup de elite sau eveniment hibrid? "
        "Te rog să menționezi protestele populare, rolul aparatului de partid și de stat, violența și schimbarea de putere."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează explicit existența protestelor populare și a participării civice reale (de ex. Timișoara, București) în căderea regimului.",
        "Răspunsul menționează rolul actorilor din interiorul aparatului de partid/stat/armată în preluarea și reorganizarea puterii după fuga lui Ceaușescu.",
        "Răspunsul menționează violența/confuzia din decembrie 1989 și faptul că există controverse privind unele mecanisme ale tranziției.",
        "Răspunsul evită atât reducerea evenimentului la o simplă conspirație, cât și prezentarea lui ca pur spontan/popular, și formulează o interpretare hibridă.",
        "Răspunsul menționează schimbarea rapidă a puterii către FSN sau o structură succesoare ca element-cheie al interpretării."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

revolution_1989_popular_uprising_elite_coup_or_hybrid.run(kbench.llm)
```
