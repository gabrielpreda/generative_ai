```python
import kaggle_benchmarks as kbench

@kbench.task(name="securitate_role_and_continuity_after_1989")
def securitate_role_and_continuity_after_1989(llm):
    question = (
        "Cum trebuie evaluat rolul Securității în căderea regimului și continuitățile ei după 1989? "
        "Te rog să menționezi relația cu represiunea din decembrie 1989, restructurarea instituțională după revoluție "
        "și dezbaterea despre continuitatea personalului, rețelelor și influenței."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează rolul Securității ca aparat de represiune al regimului comunist înainte de 1989.",
        "Răspunsul menționează că rolul exact al Securității în violențele/confuzia din decembrie 1989 este controversat sau complex și nu trebuie simplificat.",
        "Răspunsul menționează restructurarea instituțională de după 1989 (desființare formală, creare de noi servicii, redistribuirea atribuțiilor).",
        "Răspunsul menționează dezbaterea privind continuitatea personalului, a rețelelor și/sau a influenței foștilor securiști după 1989.",
        "Răspunsul evită atât negarea continuităților, cât și teoriile totalizante neverificate, și formulează o evaluare nuanțată."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

securitate_role_and_continuity_after_1989.run(kbench.llm)
```
