```python
import kaggle_benchmarks as kbench

@kbench.task(name="daco_roman_continuity_vs_migrationism")
def daco_roman_continuity_vs_migrationism(llm):
    question = (
        "În dezbaterea despre etnogeneza românilor (continuitatea daco-romană vs. teoriile migraționiste), "
        "care sunt principalele poziții și ce fel de dovezi invocă fiecare? "
        "Te rog să menționezi explicit ipoteza formării predominant la nord vs. la sud de Dunăre și să subliniezi ce este disputat."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul descrie poziția continuității daco-romane la nord de Dunăre (persistență/continuitate după retragerea romană).",
        "Răspunsul descrie poziția migraționistă / sud-dunăreană (formare predominant la sud de Dunăre și revenire/expansiune spre nord ulterior).",
        "Răspunsul menționează tipuri de dovezi folosite în dezbatere (de ex. lingvistice, arheologice, toponimice/hidronimice, surse scrise).",
        "Răspunsul semnalează explicit că problema nu are consens total și că există incertitudini/limite ale surselor pentru secolele IV–X."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total

        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

daco_roman_continuity_vs_migrationism.run(kbench.llm)
```
