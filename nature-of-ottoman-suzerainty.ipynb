```python
import kaggle_benchmarks as kbench

@kbench.task(name="nature_of_ottoman_suzerainty_vassalage_autonomy_semi_independence")
def nature_of_ottoman_suzerainty_vassalage_autonomy_semi_independence(llm):
    question = (
        "Care a fost natura suzeranității otomane asupra Țării Românești și Moldovei: "
        "vasalitate strictă, autonomie largă sau o formă de semi-independență? "
        "Te rog să explici mecanismele concrete (tribut, confirmarea domnului, obligații militare), "
        "dar și limitele controlului otoman."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează plata tributului (haraci) ca element central al relației cu Imperiul Otoman.",
        "Răspunsul menționează rolul Porții în confirmarea sau numirea domnitorilor, mai ales în perioadele ulterioare.",
        "Răspunsul menționează obligații militare sau diplomatice față de Imperiul Otoman.",
        "Răspunsul explică faptul că principatele și-au păstrat instituțiile interne, religia ortodoxă și o autonomie administrativă semnificativă.",
        "Răspunsul evită extremele (nu prezintă situația ca independență deplină, dar nici ca integrare completă într-o provincie otomană).",
        "Răspunsul semnalează că natura suzeranității a variat în timp (de ex. diferențe între secolele XV–XVI și perioada fanariotă)."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

nature_of_ottoman_suzerainty_vassalage_autonomy_semi_independence.run(kbench.llm)
```
