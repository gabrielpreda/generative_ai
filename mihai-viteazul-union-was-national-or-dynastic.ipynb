```python
import kaggle_benchmarks as kbench

@kbench.task(name="mihai_viteazul_1600_union_national_or_dynastic_project")
def mihai_viteazul_1600_union_national_or_dynastic_project(llm):
    question = (
        "Unirea realizată de Mihai Viteazul în 1600 trebuie interpretată ca un proiect de unificare națională "
        "sau ca o construcție dinastică/personală în contextul politic al epocii? "
        "Te rog să explici contextul internațional și limitele acestei uniri."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează faptul că unirea din 1600 a fost de scurtă durată și dependentă de context militar.",
        "Răspunsul explică interpretarea modernă-națională (prefigurare a unității românești).",
        "Răspunsul explică interpretarea dinastică/personală (proiect de putere în cadrul conflictelor habsburgice și antiotomane).",
        "Răspunsul menționează contextul geopolitic (Imperiul Otoman, Habsburgii, Polonia, Transilvania).",
        "Răspunsul evită anacronismul și subliniază diferența dintre conceptul modern de națiune și realitatea politică a secolului XVII."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

mihai_viteazul_1600_union_national_or_dynastic_project.run(kbench.llm)
```
