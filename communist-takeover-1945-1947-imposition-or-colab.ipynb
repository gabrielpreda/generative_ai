```python
import kaggle_benchmarks as kbench

@kbench.task(name="communist_takeover_1945_1947_soviet_imposition_vs_internal_collaboration")
def communist_takeover_1945_1947_soviet_imposition_vs_internal_collaboration(llm):
    question = (
        "Instalarea regimului comunist în România între 1945 și 1947 trebuie explicată mai ales prin "
        "impunere sovietică sau prin colaborare internă a unor elite, instituții și grupuri politice? "
        "Te rog să menționezi rolul Armatei Roșii, al PCR, al monarhiei, al administrației și pașii-cheie ai preluării puterii."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează rolul decisiv al prezenței sovietice / Armatei Roșii în schimbarea raportului de forțe după 1944.",
        "Răspunsul menționează rolul PCR, deși inițial redus ca bază socială, în capturarea treptată a instituțiilor statului.",
        "Răspunsul menționează existența colaborării interne / a actorilor locali (politicieni, funcționari, aparatul de stat, oportuniști) care au facilitat consolidarea regimului.",
        "Răspunsul menționează pași-cheie precum guvernul Groza, falsificarea/contestarea alegerilor din 1946 și abdicarea forțată a regelui în 1947.",
        "Răspunsul evită explicația exclusivă și formulează o interpretare mixtă: impunere sovietică fundamentală, dar realizată prin colaborare și adaptare internă."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

communist_takeover_1945_1947_soviet_imposition_vs_internal_collaboration.run(kbench.llm)
```
