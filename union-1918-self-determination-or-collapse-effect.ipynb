```python
import kaggle_benchmarks as kbench

@kbench.task(name="great_union_1918_self_determination_opportunity_or_collapse_effect")
def great_union_1918_self_determination_opportunity_or_collapse_effect(llm):
    question = (
        "Unirea din 1918 ('Marea Unire') trebuie interpretată în primul rând ca "
        "autodeterminare națională, oportunitate geopolitică sau ca efect al prăbușirii imperiilor "
        "(Țarist/Habsburgic/Otoman)? Te rog să explici rolul deciziilor locale, al contextului internațional "
        "și al factorilor militari/diplomatici."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează explicit prăbușirea/colapsul imperiilor (cel puțin cel Țarist și Austro-Ungar) ca factor-cheie care a deschis oportunitatea unirilor.",
        "Răspunsul menționează componenta de autodeterminare/decizii politice locale (adunări, declarații, reprezentare politică) fără a le prezenta ca singura cauză.",
        "Răspunsul menționează rolul contextului geopolitic și diplomatic de la finalul războiului (poziția Antantei, negocieri, conferințe de pace) în validarea/reconfigurarea frontierelor.",
        "Răspunsul menționează că procesul a fost diferit pe provincii (de ex. Basarabia vs Bucovina vs Transilvania) sau indică faptul că au existat dinamici regionale distincte.",
        "Răspunsul evită teleologia (nu prezintă unirea ca inevitabilă) și formulează o explicație mixtă: inițiativă politică + oportunitate strategică + colaps imperial."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

great_union_1918_self_determination_opportunity_or_collapse_effect.run(kbench.llm)
```
