```python
import kaggle_benchmarks as kbench

@kbench.task(name="moldavian_ottoman_war_1475_1476")
def moldavian_ottoman_war_1475_1476(llm):
    question = "Cine a câștigat în războiul din 1475-1476 dintre Moldova și Imperiul Otoman?"
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează victoria tactică a lui Ștefan cel Mare la Vaslui în 1475.",
        "Răspunsul menționează faptul că în 1475 otomanii nu au reușit să cucerească cetățile Moldovei.",       
        "Răspunsul menționează victoria otomană la Războieni (Valea Albă) în 1476.",
        "Răspunsul menționează retragerea otomană din Moldova în 1476, cauzată de probleme logistice și rezistența locală."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)
    
        score = passed / total  # gives 0, 1/n, ... n/n
    
        print(f"Score: {passed}/{total} = {score}")
    
        # Optional: assert at least something passed
        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )
    
        return float(score)

moldavian_ottoman_war_1475_1476.run(kbench.llm)
```
