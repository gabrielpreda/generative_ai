```python
import kaggle_benchmarks as kbench

@kbench.task(name="people_on_the_train")
def people_on_the_train(llm):
    prompt = """
    There were some people on a train.
    19 people get off the train at the first stop.
    17 people get on the train.
    Now there are 63 people on the train.
    How many people were on the train to begin with? 
    Respond with only the integer number.
"""
    initial_count = llm.prompt(prompt, schema=int)

    kbench.assertions.assert_equal(
        initial_count,
        65,
        expectation="The initial number of people on the train should be 65."
    )

people_on_the_train.run(kbench.llm)
```
