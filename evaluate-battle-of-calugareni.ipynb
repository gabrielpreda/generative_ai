```python
import kaggle_benchmarks as kbench

@kbench.task(name="evaluate_battle_of_calugareni")
def evaluate_battle_of_calugareni(llm, judge_llm):
    prompt = "Cine a câștigat bătălia de la Călugăreni?"
    response = llm.prompt(prompt)

    criteria = [
        "Răspunsul precizează victoria tactică a lui Mihai Viteazul.",
        "Răspunsul menționează succesul operațional ulterior al lui Sinan Pașa.",
        "Răspunsul menționează că Sinan Pașa a ocupat Bucureștiul și Târgoviștea.",
        "Răspunsul clarifică victoria finală a coaliției anti-otomane în campania respectivă.",
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM should have provided an assessment."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)
    
        score = passed / total  # gives 0, 1/n, ... n/n
    
        print(f"Score: {passed}/{total} = {score}")
    
        # Optional: assert at least something passed
        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )
    
        return float(score)

evaluate_battle_of_calugareni.run(llm=kbench.llm, judge_llm=kbench.judge_llm)
```
