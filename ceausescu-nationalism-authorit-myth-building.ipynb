```python
import kaggle_benchmarks as kbench

@kbench.task(name="ceausescu_nationalism_resistance_to_moscow_or_authoritarian_myth_building")
def ceausescu_nationalism_resistance_to_moscow_or_authoritarian_myth_building(llm):
    question = (
        "Naționalismul lui Ceaușescu trebuie interpretat ca o formă de rezistență față de Moscova "
        "sau mai ales ca un instrument de legitimare autoritară și mitologizare a istoriei? "
        "Te rog să menționezi politica externă, propaganda istorică și cultul personalității."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează elemente de relativă autonomie față de Moscova / URSS în anumite momente sau pe anumite dosare de politică externă.",
        "Răspunsul menționează explicit utilizarea naționalismului ca instrument de legitimare internă a regimului.",
        "Răspunsul menționează propaganda istorică / protocronismul / reinterpretarea trecutului în cheie eroică-națională.",
        "Răspunsul menționează cultul personalității lui Ceaușescu ca parte a construcției ideologice.",
        "Răspunsul evită idealizarea și formulează o interpretare mixtă, dar subliniază că autonomia externă limitată a coexistat cu un amplu proiect autoritar de mit-building."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

ceausescu_nationalism_resistance_to_moscow_or_authoritarian_myth_building.run(kbench.llm)
```
