```python
import kaggle_benchmarks as kbench

@kbench.task(name="august_23_1944_national_salvation_or_political_opportunism")
def august_23_1944_national_salvation_or_political_opportunism(llm):
    question = (
        "Actul de la 23 august 1944 trebuie interpretat ca salvare națională "
        "(ieșirea din alianța cu Germania și reducerea pierderilor) sau ca oportunism politic "
        "al elitelor și al regelui în contextul prăbușirii frontului? "
        "Te rog să menționezi contextul militar, rolul regelui Mihai, al partidelor și consecințele imediate."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează schimbarea de tabără a României în august 1944 și ruperea alianței cu Germania nazistă.",
        "Răspunsul menționează contextul militar critic (ofensiva sovietică, prăbușirea frontului, apropierea Armatei Roșii) ca element esențial.",
        "Răspunsul menționează rolul regelui Mihai și/sau al partidelor/oponenților lui Antonescu în organizarea actului de la 23 august.",
        "Răspunsul explică argumentul 'salvării naționale' (scurtarea războiului, evitarea unor distrugeri suplimentare, facilitarea ieșirii din Axă).",
        "Răspunsul explică argumentul 'oportunismului politic' (decizie luată sub presiunea contextului militar și pentru repoziționare politică).",
        "Răspunsul evită un verdict unilateral și formulează o interpretare mixtă: decizie cu efect salvator, dar luată într-un context de constrângere și calcul politic."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

august_23_1944_national_salvation_or_political_opportunism.run(kbench.llm)
```
