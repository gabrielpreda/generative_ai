```python
import kaggle_benchmarks as kbench

@kbench.task(name="slavic_influence_in_early_romanian_formation")
def slavic_influence_in_early_romanian_formation(llm):
    question = (
        "Care a fost influența slavă în formarea timpurie a românilor și a instituțiilor medievale "
        "din spațiul românesc? Explică pe scurt ponderea influenței (lingvistică, religioasă, politică) "
        "și ce arată dovezile."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează influența slavă asupra limbii române (lexic slav semnificativ și/sau terminologie religioasă/administrativă).",
        "Răspunsul menționează rolul slavonei ca limbă de cult și/sau de cancelarie în Evul Mediu.",
        "Răspunsul menționează influențe instituționale/culturale (de ex. creștinarea în rit bizantin prin mediere sud-slavă, modele de titulatură sau practici de cancelarie).",
        "Răspunsul evită extremele (nu afirmă că românii sunt „slavi”, dar nici că influența slavă e neglijabilă) și descrie un model de interacțiune/stratificare.",
        "Răspunsul face referire la tipuri de dovezi (lingvistică istorică, onomastică, toponimie, surse documentare/cancelarie, arheologie) și semnalează limitele interpretării."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

slavic_influence_in_early_romanian_formation.run(kbench.llm)
```
