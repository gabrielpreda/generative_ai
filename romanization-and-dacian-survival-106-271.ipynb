```python
import kaggle_benchmarks as kbench

@kbench.task(name="romanization_and_dacian_survival_106_271")
def romanization_and_dacian_survival_106_271(llm):
    question = (
        "Care este rolul romanizării în formarea populației romanice din spațiul carpato-danubian "
        "și cât de mare ar fi putut fi supraviețuirea demografică a dacilor după 106 până la 271? "
        "Te rog să explici pe scurt ce știm, ce nu știm, și ce indicii folosesc istoricii/arheologii."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează că romanizarea a fost un proces (administrativ, urban, economic, militar, cultural/lingvistic), nu un eveniment instantaneu.",
        "Răspunsul discută ideea de supraviețuire/continuitate a populației locale (dacice) într-o formă amestecată cu coloniști și structuri romane, fără a pretinde cifre certe.",
        "Răspunsul menționează retragerea administrației romane (271/275) și explică faptul că aceasta nu implică automat depopulare totală.",
        "Răspunsul indică tipuri de dovezi folosite pentru a discuta romanizarea și demografia (arheologie, epigrafie, toponimie/hidronimie, lingvistică, comparație regională, surse scrise).",
        "Răspunsul semnalează explicit limitele datelor și caracterul disputat al estimărilor demografice pentru 106–271."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total

        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

romanization_and_dacian_survival_106_271.run(kbench.llm)
```
