```python
import kaggle_benchmarks as kbench

@kbench.task(name="battle_of_posada_1330_scale_location_significance")
def battle_of_posada_1330_scale_location_significance(llm):
    question = (
        "Ce se știe (și ce este disputat) despre Bătălia de la Posada (1330): "
        "scara confruntării, posibila locație și semnificația ei politică? "
        "Te rog să separi ce e consensual de ce e controversat."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează consensul de bază: înfrângerea armatei regelui Carol Robert de către Basarab I într-o ambuscadă într-un defileu/trecătoare (1330).",
        "Răspunsul menționează explicit că locația exactă a 'Posadei' este disputată și că există mai multe ipoteze (fără a prezenta una drept certă).",
        "Răspunsul menționează o sursă-cheie de interpretare (de ex. Cronica pictată de la Viena / Chronicon Pictum) sau, mai general, că sursele sunt în principal maghiare și limitate.",
        "Răspunsul discută scara/cifrele cu prudență (evită numere precise prezentate ca certe; recunoaște tendința de exagerare în tradiția populară/manuale).",
        "Răspunsul explică semnificația: consolidarea autonomiei/independenței de facto a Țării Românești față de Regatul Ungariei și afirmarea dinastiei Basarabilor."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

battle_of_posada_1330_scale_location_significance.run(kbench.llm)
```
