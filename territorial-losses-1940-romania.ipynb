```python
import kaggle_benchmarks as kbench

@kbench.task(name="territorial_losses_1940_isolation_or_systemic_failure")
def territorial_losses_1940_isolation_or_systemic_failure(llm):
    question = (
        "Pierderile teritoriale ale României din 1940 trebuie explicate mai ales prin izolare diplomatică "
        "în contextul european (prăbușirea sistemului de alianțe, pacte și presiuni externe) sau prin eșecuri sistemice interne "
        "(politice, militare, instituționale)? Te rog să menționezi factorii externi și interni și să explici "
        "de ce opțiunile României au fost limitate."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează explicit contextul geopolitic al anului 1940 (al doilea război mondial și schimbarea rapidă a balanței de putere în Europa).",
        "Răspunsul menționează izolarea diplomatică / slăbirea garanțiilor și a alianțelor (de ex. prăbușirea Franței și vulnerabilizarea sistemului regional).",
        "Răspunsul menționează presiunile marilor puteri și/sau înțelegerile dintre ele (de ex. pacte/sfere de influență) ca factori externi determinanți.",
        "Răspunsul menționează factori interni: criză politică, probleme de coeziune internă, pregătire militară insuficientă sau incapacitate instituțională de a gestiona criza.",
        "Răspunsul evită o explicație unică și prezintă interacțiunea dintre constrângeri externe și vulnerabilități interne."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

territorial_losses_1940_isolation_or_systemic_failure.run(kbench.llm)
```
