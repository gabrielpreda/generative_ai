```python
import kaggle_benchmarks as kbench

@kbench.task(name="union_1859_popular_will_or_elite_maneuver")
def union_1859_popular_will_or_elite_maneuver(llm):
    question = (
        "Unirea din 1859 (dubla alegere a lui Alexandru Ioan Cuza) trebuie interpretată ca expresie a "
        "voinței naționale populare sau mai degrabă ca un rezultat al manevrelor diplomatice și de elită? "
        "Te rog să explici factorii interni (mobilizare publică, elite politice) și externi (rolul Marilor Puteri) "
        "și limitele legitimității populare în contextul epocii."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează dubla alegere a lui Alexandru Ioan Cuza în 1859 (Moldova + Țara Românească) ca mecanism central al unirii.",
        "Răspunsul explică rolul elitelor politice locale (boieri, lideri urbani, notabili) în promovarea și realizarea unirii.",
        "Răspunsul discută existența unei susțineri populare/angajări civice (susținere publică, societăți politice, presă) fără a o supraestima ca mobilizare de masă modernă.",
        "Răspunsul menționează constrângerile diplomatice și atitudinea Marilor Puteri (Franța, Rusia, Austro-Ungaria, Imperiul Otoman) în procesul de recunoaștere și în echilibrul regional.",
        "Răspunsul evită anacronisme (nu proiectează un naționalism de tip modern exclusiv popular) și explică că evenimentul a combinat voință locală + acțiune elitară + oportunități diplomatice."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

union_1859_popular_will_or_elite_maneuver.run(kbench.llm)
```
