```python
import kaggle_benchmarks as kbench

@kbench.task(name="cuman_steppe_influences_medieval_elites_basarabs")
def cuman_steppe_influences_medieval_elites_basarabs(llm):
    question = (
        "Ce rol au avut cumanii și alte influențe de stepă (turcice) în formarea elitelor medievale "
        "din Țara Românească și în special în jurul dinastiei Basarabilor? "
        "Te rog să explici ce indicii există (onomastică, context politic, surse) și ce rămâne disputat."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează explicit că numele „Basarab” este frecvent interpretat ca având origine turcică/cumană (onomastică).",
        "Răspunsul explică contextul istoric al prezenței/rolului cumanilor în spațiul dintre Carpați și Dunăre (secolele XI–XIII) și integrarea lor ulterioară.",
        "Răspunsul distinge între origine/elemente dinastice sau de elită și compoziția populației (evită concluzia simplistă că 'românii erau cumani').",
        "Răspunsul menționează tipuri de dovezi folosite: onomastică, toponimie, izvoare medievale (maghiare/latine/cronici), arheologie sau analogii regionale.",
        "Răspunsul semnalează explicit că gradul exact al contribuției demografice/etnice este disputat și că dovezile sunt indirecte."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

cuman_steppe_influences_medieval_elites_basarabs.run(kbench.llm)
```
