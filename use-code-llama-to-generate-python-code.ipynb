# Introduction  


Code generation is a very useful feature that LLMs can be used for.  
In this Notebook we will test one of the available models on Kaggle for code generation.  


Let's check the **Code LLama** for code generation.  

The model details are: 

* **Model**: Code Llama 2
* **Variation**: 7b-python-hf (7b: 7B dimm. hf: HuggingFace build)   
* **Version**: V1  
* **Framework**: PyTorch



# Import packages

```python
from transformers import AutoTokenizer
import transformers
import torch
from time import time
```

# Prepare the tokenizer and pipeline

```python
model = '/kaggle/input/codellama/pytorch/7b-python-hf/1'

time_1 = time()
tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
print(f"Tokenizer & pipeline: {round(time() - time_1)} sec.")
```

# Test with a code prompt

```python
prompt = 'Write the code for a function to compute the area of circle.'
```

```python
sequences = pipeline(
    prompt,
    do_sample=True,
    top_k=10,
    temperature=0.1,
    top_p=0.95,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=200,
)
```

```python
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```

```python
prompt = 'from sklearn.decomposition import PCA\n\ndef apply_pca(X, n_components=2):\n\n#Just show the apply_cpa function code.'
```

```python
sequences = pipeline(
    prompt,
    do_sample=True,
    top_k=10,
    temperature=0.01,
    top_p=0.95,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=256,
)
```

```python
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```
