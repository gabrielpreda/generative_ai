```python
import kaggle_benchmarks as kbench

@kbench.task(name="post_1989_transition_rupture_or_negotiated_continuity")
def post_1989_transition_rupture_or_negotiated_continuity(llm):
    question = (
        "Tranziția post-1989 în România trebuie interpretată ca o ruptură clară cu comunismul "
        "sau ca o continuitate negociată a elitelor și instituțiilor? "
        "Te rog să menționezi schimbările politice și economice, dar și continuitățile de personal, rețele și practici."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează elemente de ruptură reală după 1989 (pluralism politic, alegeri, economie de piață, libertăți civile sau orientare euro-atlantică).",
        "Răspunsul menționează continuități de elite/personal/rețele provenite din vechiul regim sau adaptarea acestora la noul context.",
        "Răspunsul menționează dificultățile și compromisurile tranziției (privatizare, instituții slabe, reforme graduale, conflict social/politic).",
        "Răspunsul evită atât teza unei rupturi totale, cât și teza unei continuități totale și formulează o interpretare mixtă de 'continuitate negociată'.",
        "Răspunsul menționează că ritmul și profunzimea rupturii au variat în timp (anii 1990 vs etape ulterioare) sau indică caracterul procesual al tranziției."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

post_1989_transition_rupture_or_negotiated_continuity.run(kbench.llm)
```
