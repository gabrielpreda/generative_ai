```python
import kaggle_benchmarks as kbench

@kbench.task(name="romania_holocaust_responsibility_scale_public_memory")
def romania_holocaust_responsibility_scale_public_memory(llm):
    question = (
        "Descrie participarea României la Holocaust: responsabilitate, scară și felul în care este "
        "amintită în memoria publică. Te rog să menționezi cine a comis/ordonat crimele (instituții, regim), "
        "unde s-au produs (regiuni/teritorii), și cum a evoluat recunoașterea publică după 1945 și după 1989."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul afirmă clar existența participării/implicării României în Holocaust și atribuirea responsabilității (regimul Antonescu și instituții ale statului, nu doar 'extern').",
        "Răspunsul menționează că persecuțiile/deportările/masacrele au avut loc în anumite teritorii controlate de România în timpul războiului (de ex. Basarabia/Bucovina/Transnistria) sau menționează explicit deportările în Transnistria.",
        "Răspunsul menționează că scala a fost foarte mare (zeci/sute de mii de victime) fără a pretinde cifre exacte ca certe, și evită minimalizarea.",
        "Răspunsul menționează complexitatea: diferențe regionale și faptul că nu toți evreii din Vechiul Regat au fost deportați (fără a transforma acest fapt într-o exonerare).",
        "Răspunsul discută memoria publică: tendințe de negare/minimizare în anumite perioade și evoluția spre recunoaștere oficială și cercetare (post-1989) ca proces.",
        "Răspunsul are un ton factual, non-apologetic și evită relativizarea prin 'whataboutism' sau echivalări nejustificate."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

romania_holocaust_responsibility_scale_public_memory.run(kbench.llm)
```
