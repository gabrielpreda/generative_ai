```python
import kaggle_benchmarks as kbench

@kbench.task(name="cheryl_birthday_puzzle")
def solve_cheryl_birthday(llm):
    prompt = """
        Solve the following logic puzzle and provide only the final date as the answer.
    
        Albert and Bernard have just become friends with Cheryl, and they want to know when her birthday is.
    
        Cheryl gives them a list of 10 possible dates:
        - May 15, May 16, May 19
        - June 17, June 18
        - July 14, July 16
        - August 14, August 15, August 17
    
        Cheryl then tells Albert the month of her birthday, and to Bernard the day of her birthday.
    
        The following conversation takes place:
        Albert: "I dont know when Cheryls birthday is, but I know that Bernard doesnt know either."
        Bernard: "At first I didnt know when Cheryls birthday was, but now I know."
        Albert: "Then I know it too."
    
        What is Cheryl's birthday?
    """
    response = llm.prompt(prompt)

    kbench.assertions.assert_contains_regex(
        r"(?i)July\s*16",
        response,
        expectation="The model should correctly deduce that Cheryl's birthday is July 16."
    )

solve_cheryl_birthday.run(kbench.llm)
```
