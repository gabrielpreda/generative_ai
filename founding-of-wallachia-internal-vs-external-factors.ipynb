```python
import kaggle_benchmarks as kbench

@kbench.task(name="founding_of_wallachia_internal_vs_hungarian_frontier_politics")
def founding_of_wallachia_internal_vs_hungarian_frontier_politics(llm):
    question = (
        "Cum explică istoriografia formarea Țării Românești (fondarea statului muntean): "
        "ca evoluție internă a formațiunilor locale vs. ca rezultat al politicilor de frontieră ale Regatului Ungariei? "
        "Menționează ce indicii există pentru fiecare (surse, contexte) și ce rămâne disputat."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul descrie perspectiva 'evoluției interne' (consolidare treptată a voievodatelor/cnezatelor locale și coagulare politică).",
        "Răspunsul descrie perspectiva 'frontierei ungare' (rolul presiunii/organizării de graniță, relațiile de vasalitate/conflict cu coroana maghiară, contextul severin/banate/mărci).",
        "Răspunsul menționează explicit figura lui Basarab I ca actor-cheie în cristalizarea statului și în relația conflictuală cu Carol Robert.",
        "Răspunsul explică că cele două perspective nu se exclud complet (interacțiune între dinamica locală și politica regională maghiară).",
        "Răspunsul invocă tipuri de dovezi (diplome/acte regale maghiare, cronici, arheologie/organizare politică, onomastică/toponimie) și semnalează limitele surselor."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

founding_of_wallachia_internal_vs_hungarian_frontier_politics.run(kbench.llm)
```
