```python
import kaggle_benchmarks as kbench

@kbench.task(name="tudor_vladimirescu_1821_social_revolution_national_or_power_struggle")
def tudor_vladimirescu_1821_social_revolution_national_or_power_struggle(llm):
    question = (
        "Cum trebuie interpretată mișcarea din 1821 condusă de Tudor Vladimirescu în Țara Românească: "
        "revoltă socială, 'trezire' națională sau o luptă pentru putere în contextul epocii? "
        "Te rog să menționezi revendicările (de ex. față de administrația fanariotă), relația cu Eteria "
        "și rezultatele imediate."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează componenta socială/politică internă (nemulțumiri față de fiscalitate, abuzuri, administrația fanariotă, boieri vs. 'norod').",
        "Răspunsul menționează componenta de afirmare politică locală (limitarea dominației fanariote/creșterea autonomiei) fără a proiecta anacronic naționalism modern deplin.",
        "Răspunsul menționează explicit relația/tensiunea dintre Tudor Vladimirescu și Eteria (Filiki Eteria) / planurile grecești antiotomane.",
        "Răspunsul menționează faptul că mișcarea s-a desfășurat într-un context regional mai larg (Imperiul Otoman, Rusia, revolta eteristă în Balcani) și că a implicat calcule de putere.",
        "Răspunsul descrie rezultatele imediate: căderea regimului fanariot în Țara Românească (și apoi și în Moldova) și revenirea la domni pământeni, chiar dacă pe termen scurt a existat represie/instabilitate.",
        "Răspunsul prezintă interpretarea ca fenomen mixt (social + politic + context geopolitic), evitând o explicație unică și unilaterală."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

tudor_vladimirescu_1821_social_revolution_national_or_power_struggle.run(kbench.llm)
```
