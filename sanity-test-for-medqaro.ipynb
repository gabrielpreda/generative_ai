# Statistics

## Length of texts

Let's check the length of texts in this dataset.

```python
val_df.columns
```

```python
for col in ['Epicriza', 'Intrebare', 'Raspuns']:
    val_df[f"text_length_{col}"] = val_df[col].str.len()
```

## Number of words

And let's check the number of words.

```python
for col in ['Epicriza', 'Intrebare', 'Raspuns']:
    val_df[f"word_count_{col}"] = val_df[col].str.split().str.len()
```

## Number of tokens

And then let's check the number of tokens.

```python
encoding = tiktoken.get_encoding("cl100k_base")

for col in ['Epicriza', 'Intrebare', 'Raspuns']:
    val_df[f"token_count_{col}"] = val_df[col].apply(lambda x: len(encoding.encode(str(x))))
```

## Summary statistics values

Now, let's calculate few stats. We will apply `describe` to the values that we just calculated, to generate the count, min, max, mean, std, and the other values of descriptive statistics.

```python
val_df.columns
```

```python
sel_columns = ['text_length_Epicriza', 'text_length_Intrebare', 'text_length_Raspuns',
       'word_count_Epicriza', 'word_count_Intrebare', 'word_count_Raspuns',
       'token_count_Epicriza', 'token_count_Intrebare', 'token_count_Raspuns']
```

```python
val_df[sel_columns].head()
```

```python
stats = []
for col in sel_columns:
    stats.append((col, val_df[col].describe().to_json()))
```

Let's put now the data in a pandas DataFrame.

```python
rows = []
for name, stats_json in stats:
    stats_dict = json.loads(stats_json)
    stats_dict["metric"] = name
    rows.append(stats_dict)

stats_df = pd.DataFrame(rows)

stats_df = stats_df.set_index("metric")
```

Let's round the values to 2 decimals, for easier readability.

```python
# Convert numeric columns
for col in stats_df.columns:
    stats_df[col] = stats_df[col].round(2)   # keep 2 decimals
```

It is much easier to see the results if be transpose this DataFrame.

```python
stats_df.T
```
