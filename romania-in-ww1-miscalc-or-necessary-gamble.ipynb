```python
import kaggle_benchmarks as kbench

@kbench.task(name="romania_ww1_1916_1918_miscalculation_or_necessary_gamble")
def romania_ww1_1916_1918_miscalculation_or_necessary_gamble(llm):
    question = (
        "România în Primul Război Mondial (1916–1918): a fost intrarea în război o eroare strategică "
        "sau un pariu necesar? Te rog să explici motivele intrării în 1916, "
        "ce s-a întâmplat militar (1916–1917), rolul Rusiei și schimbarea situației în 1918."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează motivele intrării României în 1916 (obiective teritoriale/naționale și alegerea taberei Antantei în context diplomatic).",
        "Răspunsul menționează dificultățile/campania din 1916 (înfrângeri, pierderea Bucureștiului și retragerea în Moldova) ca argument pentru 'micalculare' sau subestimare.",
        "Răspunsul menționează rezistența și reorganizarea din 1917 (bătălii defensive, refacerea armatei, sprijin aliat) ca element de evaluare nuanțată.",
        "Răspunsul menționează explicit rolul Rusiei și impactul Revoluției ruse asupra frontului (colaps/ieșire din război) și consecințele pentru România.",
        "Răspunsul menționează schimbarea din 1918 (Pacea de la București/ieșirea temporară și apoi reintrarea în finalul războiului) sau, cel puțin, faptul că 1918 a schimbat complet contextul.",
        "Răspunsul evită o concluzie unilaterală și formulează interpretarea ca balanță între risc strategic și oportunitate politică necesară în contextul epocii."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

romania_ww1_1916_1918_miscalculation_or_necessary_gamble.run(kbench.llm)
```
