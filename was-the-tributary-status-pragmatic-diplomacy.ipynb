```python
import kaggle_benchmarks as kbench

@kbench.task(name="tributary_status_pragmatic_diplomacy_or_loss_of_sovereignty")
def tributary_status_pragmatic_diplomacy_or_loss_of_sovereignty(llm):
    question = (
        "Statutul de tribut al Țării Românești și Moldovei față de Imperiul Otoman "
        "trebuie interpretat ca o formă de diplomație pragmatică pentru conservarea autonomiei "
        "sau ca o pierdere reală a suveranității? Te rog să explici mecanismele concrete și "
        "să menționezi argumentele ambelor interpretări."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul menționează plata tributului (haraci) și obligațiile politice/militare față de Poartă.",
        "Răspunsul explică faptul că principatele și-au păstrat instituții interne, elite locale și religia ortodoxă.",
        "Răspunsul prezintă interpretarea 'diplomației pragmatice' (menținerea autonomiei prin compromis).",
        "Răspunsul prezintă interpretarea 'pierderii suveranității' (limitarea politicii externe, intervenția Porții în numirea domnilor).",
        "Răspunsul evită extremele și subliniază variația în timp a gradului de control otoman."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

tributary_status_pragmatic_diplomacy_or_loss_of_sovereignty.run(kbench.llm)
```
