```python
import kaggle_benchmarks as kbench

@kbench.task(name="descalecat_dragos_bogdan_moldavian_state_formation")
def descalecat_dragos_bogdan_moldavian_state_formation(llm):
    question = (
        "Cum trebuie înțelese narațiunile despre 'descălecatul' Moldovei (Dragoș vs. Bogdan) "
        "și care este interpretarea istorică mai probabilă a procesului real de formare a statului moldovean? "
        "Menționează rolul Regatului Ungariei, al structurilor locale și diferența dintre legendă și reconstrucție istorică."
    )
    response = llm.prompt(question)

    criteria = [
        "Răspunsul explică diferența dintre tradiția cronicară/legendară a 'descălecatului' și analiza istorică modernă (proces treptat, nu un singur eveniment).",
        "Răspunsul menționează Dragoș ca lider asociat cu o formațiune inițială (adesea legată de o marcă/aranjament sub suzeranitate maghiară).",
        "Răspunsul menționează Bogdan I ca figură asociată cu desprinderea de suzeranitatea maghiară și consolidarea autonomiei Moldovei.",
        "Răspunsul include rolul Regatului Ungariei și al contextului de frontieră (marcă/organizare defensivă, interese regionale) în apariția Moldovei.",
        "Răspunsul menționează că sursele sunt limitate și că există incertitudini privind detaliile cronologice și mecanismele exacte ale formării statului."
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM-ul nu a reușit să furnizeze o evaluare."
        )
    else:
        total = len(assessment.results)
        passed = sum(1 for r in assessment.results if r.passed)

        score = passed / total
        print(f"Score: {passed}/{total} = {score}")

        kbench.assertions.assert_true(
            passed > 0,
            expectation="At least one criterion should be satisfied."
        )

        return float(score)

descalecat_dragos_bogdan_moldavian_state_formation.run(kbench.llm)
```
